```python
# import Libraries
import math
from matplotlib import pyplot as plt
from numpy import linspace

g = 10
Max_R = int(input('Max Distance(m): '))
u = math.sqrt(g*Max_R)
n = 11

list_x = []
list_y = []

for tt in linspace(0,80,n+1):
    tt_r = tt*2*math.pi/360
    t_max = 2*u*math.sin(tt_r)/g
    r = (t_max)+1
    x1 = [u*math.cos(tt_r)*t for t in linspace(0, r, 100)]
    y1 = [u*math.sin(tt_r)*t-g*t*t/2 for t in linspace(0, r, 100)]
    list_x.append(x1)
    list_y.append(y1)

for a,b in zip(list_x,list_y):
    plt.plot(a,b)

line_x = list(range(Max_R))
line_y = [0 for i in line_x]
plt.plot(line_x,line_y)
plt.show()

print(f'Deg: {linspace(0,90,n+1)}')
print(f'Max_H: {u*u/(2*g)}m,\nMax_R: {u*u/g}m,\nSpped: {u}m/s.')
```
